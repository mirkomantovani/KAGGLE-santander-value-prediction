```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
lgb_xgb = pd.read_csv('sub_lgb_xgb_ens.csv')
lgb_pca = pd.read_csv('sub_lgb_sum_kmean_pca.csv')
```

```python
models = lgb_xgb.copy()
models.rename(columns={'target':'LGB_XGB'}, inplace=True)
models['LGB_PCA']=lgb_pca['target']
#models['ens'] = (sub_lgb["target"] + sub_xgb["target"])/2
```

```python
models.head()
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 6)
plt.rcParams['figure.figsize'] = [30.0,16.0]
plt.rcParams['figure.dpi']=80
plt.rcParams.update({'font.size': 24})
```

```python
import matplotlib.pyplot as plt
start=1000
end=1100
yhat = models.iloc[start:end,1:3].values
x_coordinate = [ 1 * i for i in range(len(yhat)) ]
plt.plot(x_coordinate,yhat)
plt.legend(['LGB_XGB', 'LGB_PCA','ensemble'], loc='upper left')
plt.show()
```
